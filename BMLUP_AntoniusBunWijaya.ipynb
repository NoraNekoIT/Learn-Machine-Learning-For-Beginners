```python
#Augmentasi Image Data Train and Test
train_datagen = ImageDataGenerator(
                    rescale=1./255,
                    rotation_range=20,
                    horizontal_flip=True,
                    shear_range = 0.2,
                    validation_split=0.4,
                    fill_mode = 'nearest')
 
test_datagen = ImageDataGenerator(
                    rescale=1./255,
                    validation_split=0.4,
                    )
```

```python
#object image data generator train and validation
train_generator = train_datagen.flow_from_directory(
        base_dir,
        classes=['paper', 'rock', 'scissors'],
        subset='training',
        # train_dir,  # directory data train
        target_size=(150, 150),  # set resolusi all image to 150x150 piksel
        batch_size= 32, # because 3 class we use class_mode = 'categorical'
        class_mode='categorical',
        )
 
validation_generator = test_datagen.flow_from_directory(
        base_dir,
        classes=['paper', 'rock', 'scissors'],
        subset='validation',
        # validation_dir, # directory data validation
        target_size=(150, 150), # set resolusi all image to 150x150 piksel
        batch_size= 32 , # because 3 class we use class_mode = 'categorical'
        class_mode='categorical')
```

```python
#CNN KERAS model Sequential
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
# Callbacks 
class epoch_callback(tf.keras.callbacks.Callback):
  def on_epoch_end(self,epoch, logs=None):
    if(logs.get('accuracy') > 0.96 and logs.get('val_accuracy') > 0.96):
      print("Amazing we achive accuracy {:.2f} percent and val_accuracy {:.2f} percent".format(logs.get('accuracy'),logs.get('val_accuracy')))
      self.model.stop_training = True
```

```python
# compile model with 'adam' optimizer loss function 'categorical_crossentropy' 
model.compile(loss='categorical_crossentropy',
              optimizer=tf.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
# train model with model.fit 
model.fit(
      train_generator,
      steps_per_epoch=25,  #how many batches will be executed on each epoch
      epochs=20, # add eposchs if the accuracy of the model is not optimal
      validation_data=validation_generator, # display validation data test accuracy
      validation_steps=5,  # how many batches will be executed on each epoch
      verbose=2,
      callbacks = [epoch_callback()]
      )
```
